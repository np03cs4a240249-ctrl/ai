```python
# ==============================
# Problem 1B: Data Extraction & Cleaning
# ==============================

# 1. Filter dataset for years 2020, 2021, 2022
hdi_1b_df = df[df['year'].isin([2020, 2021, 2022])].copy()

# 2. Check missing values in essential columns
print("Missing values before cleaning:")
print(hdi_1b_df[['hdi', 'country', 'year']].isna().sum())

# 3. Convert numeric columns (handle text and symbols like '–')
numeric_cols = ['hdi', 'gross_inc_percap', 'life_expectancy']
for col in numeric_cols:
    hdi_1b_df[col] = pd.to_numeric(hdi_1b_df[col], errors='coerce')

# 4. Drop rows with missing essential values (EXPLICIT DROPPING)
hdi_1b_df = hdi_1b_df.dropna(subset=['hdi', 'country', 'year'])

# 5. Remove duplicate rows
hdi_1b_df = hdi_1b_df.drop_duplicates()

# 6. Fix inconsistent or misspelled country names
hdi_1b_df['country'] = hdi_1b_df['country'].str.strip().str.title()

# 7. Ensure year is integer
hdi_1b_df['year'] = hdi_1b_df['year'].astype(int)

# 8. Final check after cleaning
print("\nMissing values after cleaning:")
print(hdi_1b_df[['hdi', 'country', 'year']].isna().sum())

# 9. Save cleaned dataset
hdi_1b_df.to_csv("HDI_problem1B.csv", index=False)

print("\nData cleaning completed and file saved as HDI_problem1B.csv")
```

 PROBLEM 2 — ADVANCED HDI EXPLORATION

```python
#1. Create South Asia Subset:
south_asia_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Filter dataset for South Asia and latest year (2022)
south_asia_df = df[
    (df['country'].isin(south_asia_countries)) &
    (df['year'] == 2022)
].copy()

# Verify each country appears once
south_asia_df['country'].value_counts()
```

```python
south_asia_df.to_csv("HDI_SouthAsia.csv", index=False)
```

```python
# Convert columns to numeric
south_asia_df['life_expectancy'] = pd.to_numeric(
    south_asia_df['life_expectancy'], errors='coerce'
)

south_asia_df['gross_inc_percap'] = pd.to_numeric(
    south_asia_df['gross_inc_percap'], errors='coerce'
)
```

Composite Score =
0.30 × Life Expectancy Index + 0.30 × GNI per Capita Index

```python
south_asia_df['Composite Score'] = (
    0.30 * south_asia_df['life_expectancy'] +
    0.30 * south_asia_df['gross_inc_percap']
)

south_asia_df[['country', 'Composite Score']]
```

Rank South Asian countries based on Composite Score.

```python
composite_rank = south_asia_df.sort_values(
    by='Composite Score', ascending=False
)

composite_rank[['country', 'Composite Score']]
```

Horizontal Bar Chart — Top 5 Countries

```python
composite_rank = south_asia_df.sort_values(by='Composite Score', ascending=False)

top5 = composite_rank.head(5)

plt.figure(figsize=(8, 5))
plt.barh(top5['country'], top5['Composite Score'])
plt.xlabel("Composite Development Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Development Score (2022)")
plt.gca().invert_yaxis()
plt.show()
```

Detect Outliers (HDI & GNI)

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

south_asia_df['HDI Outlier'] = detect_outliers(south_asia_df['hdi'])
south_asia_df['GNI Outlier'] = detect_outliers(south_asia_df['gross_inc_percap'])
```

```python
plt.figure(figsize=(7, 5))

plt.scatter(
    south_asia_df['gross_inc_percap'],
    south_asia_df['hdi'],
    c=south_asia_df['HDI Outlier'],
    cmap='coolwarm'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("GNI per Capita vs HDI (South Asia, 2022)")
plt.show()
```

Correlation with HDI

```python
south_asia_df['gender_development'] = pd.to_numeric(
    south_asia_df['gender_development'], errors='coerce'
)

corr_life = south_asia_df['hdi'].corr(south_asia_df['life_expectancy'])
corr_gender = south_asia_df['hdi'].corr(south_asia_df['gender_development'])

corr_life, corr_gender
```

```python
sns.lmplot(
    data=south_asia_df,
    x='life_expectancy',
    y='hdi',
    height=5,
    aspect=1.2
)
plt.title("HDI vs Life_Expectancy (South Asia, 2022)")
plt.show()

sns.lmplot(
    data=south_asia_df,
    x='gender_development',
    y='hdi',
    height=5,
    aspect=1.2
)
plt.title("HDI vs Gender Development Index (South Asia, 2022)")
plt.show()
```

GAP ANALYSIS

```python
#Create GNI–HDI Gap
south_asia_df['GNI_HDI_Gap'] = (
    south_asia_df['gross_inc_percap'] - south_asia_df['hdi']
)

south_asia_df[['country', 'GNI_HDI_Gap']]
```

```python
gap_sorted = south_asia_df.sort_values(
    by='GNI_HDI_Gap', ascending=False
)

top_pos = gap_sorted.head(3)
top_neg = gap_sorted.tail(3)

plt.figure(figsize=(8, 5))
plt.bar(top_pos['country'], top_pos['GNI_HDI_Gap'], label="Positive Gap")
plt.bar(top_neg['country'], top_neg['GNI_HDI_Gap'], label="Negative Gap")

plt.xlabel("Country")
plt.ylabel("GNI–HDI Gap")
plt.title("Top Positive and Negative GNI–HDI Gaps (South Asia, 2022)")
plt.legend()
plt.show()
```

Problem *3*

South Asia vs Middle East (2020–2022)

```python
# Load cleaned dataset from Problem 1B
hdi_1b_df = pd.read_csv("HDI_problem1B.csv")

hdi_1b_df.head()
```

```python
south_asia_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

south_asia_3_df = hdi_1b_df[hdi_1b_df['country'].isin(south_asia_countries)].copy()
middle_east_3_df = hdi_1b_df[hdi_1b_df['country'].isin(middle_east_countries)].copy()
```

```python
south_asia_3_df.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
middle_east_3_df.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

2. Descriptive Statistics:

```python
south_stats = south_asia_3_df['hdi'].agg(['mean', 'std'])
middle_stats = middle_east_3_df['hdi'].agg(['mean', 'std'])

south_stats, middle_stats
```

```python
if south_stats['mean'] > middle_stats['mean']:
    print("South Asia has higher average HDI")
else:
    print("Middle East has higher average HDI")
```

3. Top and Bottom Performers:

```python
south_top = south_asia_3_df.sort_values('hdi', ascending=False).head(3)
south_bottom = south_asia_3_df.sort_values('hdi').head(3)

middle_top = middle_east_3_df.sort_values('hdi', ascending=False).head(3)
middle_bottom = middle_east_3_df.sort_values('hdi').head(3)
```

```python
comparison_df = pd.concat([
    south_top.assign(Region="South Asia"),
    south_bottom.assign(Region="South Asia"),
    middle_top.assign(Region="Middle East"),
    middle_bottom.assign(Region="Middle East")
])

plt.figure(figsize=(10,6))
sns.barplot(data=comparison_df, x='country', y='hdi', hue='Region')
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("HDI")
plt.title("Top and Bottom HDI Performers: South Asia vs Middle East")
plt.show()
```

4. Metric Comparisons:

```python
metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']

for metric in metrics:
    plt.figure(figsize=(6,5))
    sns.barplot(
        data=pd.concat([
            south_asia_3_df.assign(Region="South Asia"),
            middle_east_3_df.assign(Region="Middle East")
        ]),
        x='Region',
        y=metric
    )
    plt.title(f"{metric.title()} Comparison by Region")
    plt.show()
```

5. HDI Disparity:

```python
def hdi_variation(df):
    hdi_range = df['hdi'].max() - df['hdi'].min()
    cv = df['hdi'].std() / df['hdi'].mean()
    return hdi_range, cv

south_range, south_cv = hdi_variation(south_asia_3_df)
middle_range, middle_cv = hdi_variation(middle_east_3_df)

south_range, south_cv, middle_range, middle_cv
```

```python
if south_cv > middle_cv:
    print("South Asia shows greater HDI variation")
else:
    print("Middle East shows greater HDI variation")
```

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return (series < Q1 - 1.5 * IQR) | (series > Q3 + 1.5 * IQR)

south_asia_3_df['HDI Outlier'] = detect_outliers(south_asia_3_df['hdi'])
middle_east_3_df['HDI Outlier'] = detect_outliers(middle_east_3_df['hdi'])
```
