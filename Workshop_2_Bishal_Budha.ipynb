```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Bishal AI Nasty/medical_students_dataset.csv")

print(df.info())
print(df.isnull().sum())

num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print(df.duplicated().sum())
df = df.drop_duplicates()

df.head()
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Dataset/Titanic-Dataset.csv")

df_sub = df[['Name','Pclass','Sex','Age','Fare','Survived']]

df_first = df_sub[df_sub['Pclass'] == 1]

mean_fare = df_first['Fare'].mean()
median_fare = df_first['Fare'].median()
max_fare = df_first['Fare'].max()
min_fare = df_first['Fare'].min()

mean_fare, median_fare, max_fare, min_fare
```

```python
onehot = pd.get_dummies(df['Embarked'], prefix='Embarked')
df2 = df
df2 = pd.concat([df2, onehot], axis=1)

df2 = df2.drop(columns=['Embarked'])

print(df2.head())
```

```python
import matplotlib.pyplot as plt

survival_rates = df2.groupby('Sex')['Survived'].mean()
print(survival_rates)

plt.figure(figsize=(6,4))
survival_rates.plot(kind='bar')
plt.title("Mean Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
plt.show()
```

```python
df3 = pd.read_csv("/content/drive/MyDrive/Dataset/Titanic-Dataset.csv")

survival_by_port = df3.groupby('Embarked')['Survived'].mean()
print(survival_by_port)

plt.figure(figsize=(6,4))
survival_by_port.plot(kind='bar')
plt.title("Mean Survival Rate by Port of Embarkation")
plt.ylabel("Survival Rate")
plt.xlabel("Embarked (C, Q, S)")
plt.show()
```
