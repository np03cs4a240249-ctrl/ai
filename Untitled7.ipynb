```python
import pandas as pd
import matplotlib.pyplot as plt

# Daily wages and corresponding number of workers
wages = [5, 7, 8, 10, 11]
frequencies = [20, 15, 12, 15, 18]

# Create a DataFrame
df = pd.DataFrame({
    'Daily Wages (Rs)': wages,
    'No. of Workers': frequencies
})

# Calculate cumulative frequency
df['Cumulative Frequency'] = df['No. of Workers'].cumsum()

# Total number of workers
N = sum(frequencies)

# Median position
median_pos = (N + 1) / 2

# Find the median wage
median_wage = None
for i in range(len(df)):
    if df.loc[i, 'Cumulative Frequency'] >= median_pos:
        median_wage = df.loc[i, 'Daily Wages (Rs)']
        break

print("Median Daily Wage:", median_wage)

# Plotting the frequency distribution
plt.bar(df['Daily Wages (Rs)'], df['No. of Workers'], color='skyblue')
plt.xlabel('Daily Wages (Rs)')
plt.ylabel('Number of Workers')
plt.title('Frequency Distribution of Daily Wages')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np

# Step 1: Reconstruct the dataset from class intervals and frequencies
intervals = [(15, 20), (20, 25), (25, 30), (30, 35)]
frequencies = [8, 21, 15, 4]

# Use midpoints to approximate values
midpoints = [(low + high) / 2 for low, high in intervals]

# Expand the dataset
data = []
for midpoint, freq in zip(midpoints, frequencies):
    data.extend([midpoint] * freq)

# Convert to numpy array
data = np.array(data)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Step 4: Detect Outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]

# Print results
print("Range:", data_range)
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Outliers:", outliers if len(outliers) > 0 else "No outliers")
```

```python
import numpy as np

# Your dataset
data = np.array([1, 3, 4, 6, 13, 20, 25, 26, 28, 62, 95])

# Range
data_range = np.max(data) - np.min(data)

# Quartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)

# IQR
IQR = Q3 - Q1

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = data[(data < lower_bound) | (data > upper_bound)]

print("Range:", data_range)
print("Q1:", Q1)
print("Median (Q2):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
import numpy as np

# Your dataset
C = np.array([112, 100, 127, 120, 134, 118, 105, 110])

# Sample variance (ddof=1 ensures sample variance, not population variance)
sample_variance = np.var(C, ddof=1)

# Sample standard deviation
sample_std_dev = np.std(C, ddof=1)

print("Sample Variance:", sample_variance)
print("Sample Standard Deviation:", sample_std_dev)
```
