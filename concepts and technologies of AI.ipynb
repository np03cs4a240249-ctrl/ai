```python
def add_binary(a, b):
    """
    Returns the sum of two decimal numbers in binary digits.

    Parameters:
        a (int): A decimal integer
        b (int): Another decimal integer

    Returns:
        binary_sum (str): Binary string of the sum of a and b
    """
    binary_sum = bin(a + b)[2:]   # Convert sum to binary, remove '0b'
    return binary_sum

print(add_binary(2, 3))   # Output: 101
```

```python
def temperature_converter():
    """
    Converts temperature between Celsius and Fahrenheit.
    This function prompts the user to specify the conversion type 
    (Celsius to Fahrenheit or Fahrenheit to Celsius)
    and then asks for the temperature value. 
    It calculates and returns the converted temperature.

    Returns:
        float: The converted temperature value.
    """

    print("Choose conversion type:")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")

    # Get conversion choice from user
    choice = input("Enter 1 or 2: ")

    if choice == "1":
        # Celsius to Fahrenheit conversion
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius}°C is equal to {fahrenheit}°F")
        return fahrenheit

    elif choice == "2":
        # Fahrenheit to Celsius conversion
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit}°F is equal to {celsius}°C")
        return celsius

    else:
        print("Invalid choice. Please enter 1 or 2.")
        return None


# Call the function
temperature_converter()
```

```python
def temperature_converter():
    """
    Converts temperature between Celsius and Fahrenheit using try-except-finally.
    """

    try:
        print("Choose conversion type:")
        print("1. Celsius to Fahrenheit")
        print("2. Fahrenheit to Celsius")

        choice = input("Enter 1 or 2: ")

        if choice == "1":
            celsius = float(input("Enter temperature in Celsius: "))
            fahrenheit = (celsius * 9/5) + 32
            print(f"{celsius}°C is equal to {fahrenheit}°F")
            return fahrenheit

        elif choice == "2":
            fahrenheit = float(input("Enter temperature in Fahrenheit: "))
            celsius = (fahrenheit - 32) * 5/9
            print(f"{fahrenheit}°F is equal to {celsius}°C")
            return celsius

        else:
            print("Invalid choice. Please enter 1 or 2.")
            return None

    except ValueError:
        print("Error: Please enter a valid number.")
        return None

    finally:
        print("Temperature conversion process completed.")  # Always runs


# Call the function
temperature_converter()
```

```python
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
] 
# 1. Create empty classification lists
low=[] # < 3 hours
moderate=[] # 3 to 5 hours

high=[] # > 5 hours

# 2. Iterate and classify study hours

for study, entertainment, sleep in time_data:
    if study < 3:
        low.append(study)
    elif 3<= study <=5:
        moderate.append(study)
    else:
        high.append(study)

# 3. Print all categories
print("Low Study Time (<3 hours):", low)
print("Moderate Study Time (3–5 hours):", moderate)
print("High Study Time (>5 hours):", high)
```

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))  # Output: 36
```

```python
def generate_permutations(s):
    if len(s) == 1:
        return [s]

    perms = []
    for i in range(len(s)):
        char = s[i]
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            new_perm = char + perm
            if new_perm not in perms:   # avoid duplicates
                perms.append(new_perm)

    return perms

print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    total = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total += calculate_directory_size(value)
        else:
            total += value
    return total

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print(calculate_directory_size(directory_structure))  # Output: 1400
```

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1, 2, 5], 11))  # Output: 3
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)

    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[n][m]

print(longest_common_subsequence("abcde", "ace"))  # Output: 3
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        wt = weights[i-1]
        val = values[i-1]

        for cap in range(1, capacity + 1):
            if wt <= cap:
                dp[i][cap] = max(val + dp[i-1][cap - wt], dp[i-1][cap])
            else:
                dp[i][cap] = dp[i-1][cap]

    return dp[n][capacity]

print(knapsack([1, 3, 4, 5], [1, 4, 5, 7], 7))  # Output: 9
```
